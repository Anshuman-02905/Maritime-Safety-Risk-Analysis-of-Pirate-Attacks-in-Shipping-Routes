```python
def fillna_with_nearest_mean(temp_df, value_col, n=5):
    filled_df = temp_df.copy()
    for i in filled_df.index:
        if pd.isna(filled_df.loc[i, value_col]):
            # Get the indices for available values
            available_values = []
            for j in range(1, n+1):
                if i-j >= 0:
                    available_values.append(filled_df.loc[i-j, value_col])
                if i+j < len(filled_df):
                    available_values.append(filled_df.loc[i+j, value_col])
            available_values = [x for x in available_values if not pd.isna(x)]
            if len(available_values) > 0:
                filled_df.loc[i, value_col] = np.mean(available_values[:n])
    if filled_df[value_col].isnull().sum()>0:
        filled_df[value_col].fillna(filled_df[value_col].median(), inplace=True)
    return filled_df
```

```python
cols=['corruption_index', 'homicide_rate', 'GDP', 'total_fisheries_per_ton',
       'total_military', 'population', 'unemployment_rate', 'totalgr',
       'industryofgdp']
```

```python
import numpy as np
cont=country_indicatorsclean.country.unique()
country_df=pd.DataFrame(columns=country_indicatorsclean.columns)
for i in cont:
  country_temp_df=country_indicatorsclean[country_indicatorsclean['country']==i]
  country_temp_df.reset_index(drop=True,inplace=True)
  for j in cols:
    country_temp_df=fillna_with_nearest_mean(country_temp_df,j)
  country_df=pd.concat([country_df,country_temp_df],ignore_index=True)
  country_df.reset_index(drop=True,inplace=True)
temp=pd.merge(country_df,country_codes,on='country')
country_df=pd.merge(country_df,country_codes,on='country')

for i in cont:
  for j in cols:
    if country_df[country_df['country']==i][j].isnull().sum()>0:
      region=list(temp[temp['country']==i]['region'])[0]
      val=temp.groupby('region')[j].median()[region]
      country_df.loc[country_df['country'] == i, j] = country_df.loc[country_df['country'] == i, j].fillna(val)


```
